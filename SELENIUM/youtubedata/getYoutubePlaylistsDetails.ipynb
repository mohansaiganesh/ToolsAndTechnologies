```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

```python
# Function to scroll down the page
def scroll_down(driver):
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
```

```python
# Replace with the URL of the YouTube channel
channel_url = "https://www.youtube.com/@takeUforward"

# path_chrome_driver = "C:\\Users\\GANESH\\Downloads\\chromedriver\\chromedriver-win64\\chromedriver.exe"
driver = webdriver.Chrome()

# Open the channel URL
driver.get(channel_url)

#use any one of the below
# print(driver.find_element(By.CSS_SELECTOR,'#tabsContent > yt-tab-group-shape > div.yt-tab-group-shape-wiz__tabs').text)
# print(driver.find_element(By.CLASS_NAME,'yt-tab-group-shape-wiz__tabs').text)


Video_types = driver.find_element(By.CLASS_NAME,'yt-tab-group-shape-wiz__tabs').text.split('\n')
# print(Video_types)

#dictionary to store the details of all the playlists
playlists_content = {}

if 'Playlists' in Video_types:
    playlists_url = channel_url+"/playlists"
    # print(playlists_url)

    driver.get(playlists_url)

    #get all the titles of the playlists
    titles_webele = driver.find_elements(By.CSS_SELECTOR,'#video-title')

    #get all the hrefs of the playlists
    playlists_webele = driver.find_elements(By.CSS_SELECTOR,'#view-more > a')

    #get number of videos
    videos_count = driver.find_elements(By.CSS_SELECTOR,'#overlays > ytd-thumbnail-overlay-bottom-panel-renderer > yt-formatted-string ')

    #to access href 
    # playlists[i].get_attribute('href')

    playlists_details = []
    for i in range(len(titles_webele)):
        playlists_details.append([titles_webele[i].text, videos_count[i].text, playlists_webele[i].get_attribute('href')])

    # print(playlists_details)

    for title, videoCount, link in playlists_details:
        playlists_content[(title,link)] = []


    for title, videoCount, link in playlists_details:
        driver.get(link)

        video_all_titles_webele = []
        video_all_duration_webele = []
        video_all_info_webele = []

        while True:
            # Scroll down to load more videos
            scroll_down(driver)
            
            # Wait for the new elements to load
            time.sleep(1)  # Adjust sleep time as necessary


            video_titles_webele = driver.find_elements(By.CSS_SELECTOR,'#video-title')
            video_duration_webele = driver.find_elements(By.CSS_SELECTOR,'#overlays > ytd-thumbnail-overlay-time-status-renderer > div.thumbnail-overlay-badge-shape.style-scope.ytd-thumbnail-overlay-time-status-renderer > badge-shape > div')
            video_info_webele = driver.find_elements(By.CSS_SELECTOR,'#video-info')

            if len(video_titles_webele) > len(video_all_info_webele):
                video_all_titles_webele = video_titles_webele
                video_all_duration_webele = video_duration_webele
                video_all_info_webele = video_info_webele
            else:
                break

        if len(video_all_titles_webele) > len(video_all_duration_webele):
            temp = ['NA',]
            temp = temp*(len(video_all_titles_webele) - len(video_all_duration_webele))
            video_all_duration_webele.extend(temp)

        if len(video_all_titles_webele) > len(video_all_info_webele):
            temp = ['NA',]
            temp = temp*(len(video_all_titles_webele) - len(video_all_info_webele))
            video_all_info_webele.extend(temp)

        try:    
            for i in range(len(video_all_titles_webele)):

                if (type(video_all_titles_webele[i]) == str):
                    video_title = video_all_titles_webele[i]
                else:
                    video_title = video_all_titles_webele[i].text

                if (type(video_all_info_webele[i]) == str):
                    video_info = video_all_info_webele[i]
                else:
                    video_info = video_all_info_webele[i].text


                if (type(video_all_duration_webele[i]) == str):
                    video_duration = video_all_duration_webele[i]
                    print('videoDuration ',video_duration)
                else:
                    video_duration = video_all_duration_webele[i].text

                if (type(video_all_titles_webele[i]) == str):
                    video_link = "NA"
                else:
                    video_link = video_all_titles_webele[i].get_attribute('href')


                playlists_content[(title,link)].append([ video_title, video_duration, video_info, video_link ])
        except Exception as e:
            print(str(e))
        # break

# #print the details of the playlists
count = 0

for key, value in playlists_content.items():
    count=count+1
    print(f"{count}. {key[0]} --- {key[1]}")

    subcount = 0

    for video_title, video_duration,video_info,video_link in value:
        subcount=subcount+1
        print(f"\t{subcount}. {video_title} ({video_duration}) {video_info}")


    print("\n")
```
